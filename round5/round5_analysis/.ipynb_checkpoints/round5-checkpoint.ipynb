```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')


trade_files = [
    ('/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-5-island-data-bottle/trades_round_5_day_2.csv', 2),
    ('/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-5-island-data-bottle/trades_round_5_day_3.csv', 3),
    ('/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-5-island-data-bottle/trades_round_5_day_4.csv', 4),
]


# Load and preprocess each day's data
dfs = []
for filepath, day in trade_files:
    df_day = pd.read_csv(filepath, sep=';')
    df_day['absolute_timestamp'] = df_day['timestamp'] + (day - 2) * 1_000_000
    df_day['timestamp'] = df_day['absolute_timestamp']
    df_day['tick'] = df_day['absolute_timestamp'] // 100
    df_day['day'] = day
    dfs.append(df_day)

# Concatenate into a single DataFrame
combined = pd.concat(dfs, ignore_index=True)

# Display basic info and sample
print("Combined shape:", combined.shape)
combined.tail()
```

```python
# Map day index to timestamp offset
# File paths and corresponding days
price_files = [
    #'/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-1-island-data-bottle/prices_round_1_day_-2.csv',
    #'/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-2-island-data-bottle/prices_round_2_day_-1.csv',
    #'/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-3-island-data-bottle/prices_round_3_day_0.csv',
    #'/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-4-island-data-bottle/prices_round_4_day_1.csv',
    '/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-5-island-data-bottle/prices_round_5_day_2.csv',
    '/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-5-island-data-bottle/prices_round_5_day_3.csv',
    '/Users/Maël Kupperschmitt/OneDrive/Desktop/IMC/round-5-island-data-bottle/prices_round_5_day_4.csv'
]

day_offsets = {
    0: 0,
    1: 1_000_000,
    2: 2_000_000,
    #3: 3_000_000,
    #4: 4_000_000,
    #5: 5_000_000,
    #6: 6_000_000,
    #7: 7_000_000
}

dfs = []

for i, file in enumerate(price_files):
    df = pd.read_csv(file)
    df = df.iloc[:, 0].str.split(";", expand=True)
    df.columns = [
        "day", "timestamp", "product",
        "bid_price_1", "bid_volume_1", "bid_price_2", "bid_volume_2", "bid_price_3", "bid_volume_3",
        "ask_price_1", "ask_volume_1", "ask_price_2", "ask_volume_2", "ask_price_3", "ask_volume_3",
        "mid_price", "profit_and_loss"
    ]
    
    # Convert numeric columns
    numeric_cols = df.columns.difference(["day", "timestamp", "product"])
    df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
    
    # Convert and shift timestamp
    df["timestamp"] = df["timestamp"].astype(int) + day_offsets[i]
    
    dfs.append(df)

# Concatenate all days into one dataframe
df = pd.concat(dfs, ignore_index=True)
```

```python

def plot_product_trades(product_df, trade_df, product_name):
    plt.figure(figsize=(12, 6))
    
    # Plot mid price
    product_data = product_df[product_df['product'] == product_name]
    plt.plot(product_data['timestamp'], product_data['mid_price'], label=f'{product_name} Mid Price', color='black', linewidth=2)
    
    # Filter the trades related to this product
    product_trades = trade_df[trade_df['symbol'] == product_name]
    
    # Plot buys (dots) and sells (crosses)
    buyers = product_trades[product_trades['buyer'] == 'Gary']  # Example for Caesar, loop for others
    plt.scatter(buyers['timestamp'], buyers['price'], label='Gary Buys', color='blue', marker='o')

    sellers = product_trades[product_trades['seller'] == 'Gary']  # Example for Paris, loop for others
    plt.scatter(sellers['timestamp'], sellers['price'], label='Gary Sells', color='red', marker='x')

    # Add additional people as needed (e.g., loop over 'buyer' and 'seller' and add them to the plot)

    # Labels and title
    plt.title(f"Mid Price vs Trades for {product_name}")
    plt.xlabel("Timestamp")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()

products = df['product'].unique()
for product in products:
    plot_product_trades(df, combined, product)
```
